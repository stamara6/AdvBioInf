```python
reader = vcf.VCFReader(open("PIK3CA.vcf", 'r'))
d1 = []
for idx,r in enumerate(reader):
    d1.append({})   
    for s in r.samples:
        d1[idx][s.sample] = s["AD"]
print(d1)
```

```python
reader = vcf.VCFReader(open("PIK3CA.vcf", 'r'))
d2 = []
for idx,r in enumerate(reader):
    d2.append({})
    d2[idx] = {'chrom' : r.CHROM,
                   'pos' : r.POS,
                   'ref' : r.REF,
                   'alt' : "%s"%r.ALT}
    for s in r.samples:
        d2[idx]["samples"] = {}
print(d2)
```

```python
final_d = {}
for ind,d in enumerate(d2):
    d["samples"] = d1[ind]
    final_d[ind] = d
print(final_d)
```

```python
final_short_d = {k: final_d[k] for k in (0, 1, 2)}
```

```python
res = es.index(index="sem4", doc_type='tweet', id=1, body=final_short_d)
```

```python
res = es.get(index="sem4", doc_type='tweet', id=1)
print(res["_source"])
```

```python
res = es.search(index="sem4", body={"query":{"match_all": {}}})
r = json.dumps(res['hits']['hits'][0]['_source'])
r_loaded = json.loads(r)
#r_loaded["0"]["samples"]

```

```python
value = [[],[],[]]
for r in r_loaded:
    for s in r_loaded[r]["samples"]:
        a1 = r_loaded[r]["samples"][s][0]
        a2 = r_loaded[r]["samples"][s][1]
        if a2 == 0:
            aa = 0
        else:
            aa = a2/(a1+a2)
        value[int(r)].append(float(aa))
```

```python
#plot distribution per record
import matplotlib
matplotlib.pyplot.hist(value)
```
